```python
fig1 = plt.figure(figsize = (5,5))
image = pil.open('RGB.jpg').convert('L')
image = image.resize((320, 200))
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.show()
```

```python
#%matplotlib notebook #Fonctionne seulement avec jupyter notebook pour un graphique interactif
xx, yy = np.mgrid[0:image.height, 0:image.width]
pixels = np.array(image.getdata()).reshape((image.height, image.width)) 
fig2 = plt.figure(figsize = (10,10))
ax = fig2.gca(projection='3d')
ax.view_init(70, 0)
ax.plot_surface(xx, yy, pixels ,rstride=1, cstride=1, cmap=plt.cm.gray,linewidth=0)
plt.show()
```
